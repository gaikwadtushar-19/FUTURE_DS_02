```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
```

```python
df['TotalCharges'] = df['TotalCharges'].fillna(df['TotalCharges'].median())
```

```python
df['ChurnFlag'] = df['Churn'].map({'Yes': 1, 'No': 0})
```

```python
df['TenureGroup'] = pd.cut(
    df['tenure'],
    bins=[0, 12, 24, 48, 72],
    labels=['0–1 year', '1–2 years', '2–4 years', '4+ years']
)
```

```python
overall_churn = df['ChurnNum'].mean()
overall_churn
```

```python
churn_rate = df['ChurnFlag'].mean()
print(f"Overall Churn Rate: {churn_rate:.2%}")
```

```python
df.groupby('Churn')['tenure'].mean()
```

```python
df.groupby('Churn')['TotalCharges'].mean()
```

```python
df.groupby('Churn')['MonthlyCharges'].mean()
```

```python
internet_churn = df.groupby('InternetService')['ChurnFlag'].mean().reset_index()

plt.figure()
plt.bar(internet_churn['InternetService'], internet_churn['ChurnFlag'])
plt.title("Churn Rate by Internet Service")
plt.ylabel("Churn Rate")
plt.show()
```

```python
contract_churn = df.groupby('Contract')['ChurnFlag'].mean().reset_index()

plt.figure()
plt.bar(contract_churn['Contract'], contract_churn['ChurnFlag'])
plt.title("Churn Rate by Contract Type")
plt.ylabel("Churn Rate")
plt.show()
```

```python
tenure_churn = (
    df.assign(ChurnNum=df['Churn'].map({'Yes': 1, 'No': 0}))
      .groupby('TenureGroup', observed=True)['ChurnNum']
      .mean()
)
```

```python
tenure_churn = (
    df.assign(ChurnNum=df['Churn'].map({'Yes': 1, 'No': 0}))
      .groupby('TenureGroup', observed=True)['ChurnNum']
      .mean()
)

plt.figure()
tenure_churn.plot(kind='bar')
plt.title("Churn Rate by Tenure Group")
plt.ylabel("Churn Rate")
plt.xlabel("Tenure Group")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
df['ChurnNum'] = df['Churn'].map({'Yes': 1, 'No': 0})
```

```python
df[['Churn', 'ChurnNum']].head()
```

```python
contract_churn = (
    df.groupby('Contract')['ChurnNum']
      .mean()
      .sort_values(ascending=False)
)

contract_churn.plot(kind='bar')
plt.title("Churn Rate by Contract Type")
plt.ylabel("Churn Rate")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
payment_churn = (
    df.groupby('PaymentMethod')['ChurnNum']
      .mean()
      .sort_values(ascending=False)
)

payment_churn.plot(kind='bar')
plt.title("Churn Rate by Payment Method")
plt.ylabel("Churn Rate")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
internet_churn = (
    df.groupby('InternetService')['ChurnNum']
      .mean()
)

internet_churn.plot(kind='bar')
plt.title("Churn Rate by Internet Service")
plt.ylabel("Churn Rate")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
df['ChargeGroup'] = pd.cut(
    df['MonthlyCharges'],
    bins=[0, 35, 70, 100, 150],
    labels=['Low', 'Medium', 'High', 'Very High']
)

df.groupby('ChargeGroup',observed=True)['ChurnNum'].mean().plot(kind='bar')
plt.title("Churn Rate by Monthly Charge Group")
plt.ylabel("Churn Rate")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
